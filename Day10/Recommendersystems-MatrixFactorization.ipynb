```python
rating=pd.merge(movies,rating)
```

```python
mr=rating.pivot_table(index=['user_id'],columns=['title'],values='rating')
```

```python
mr.head()   # Kullanıcı hangi filme ne kadar puan vermiş onu ayarladık
```

```python
mr[['Star Wars (1977)','101 Dalmatians (1996)']].corr()   # birbiri arasında ki korelasyona bakıyoruz
```

```python
rating.movie_id.nunique()
```

```python
sm=mr.corrwith(swr)
```

```python
sm.sort_values(ascending=False)
```

```python
import numpy as np
```

```python
mstats=rating.groupby('title').agg({'rating':[np.size,np.mean]})
```

```python
mstats
```

```python
pm=mstats[mstats['rating']['size']>100]
```

```python
pm
```

```python
pm.sort_values(('rating','mean'),ascending=False)
```

```python
smdata=pd.DataFrame(sm,columns=['similarity'])
```

```python
pm.columns=pm.columns.get_level_values(0)
df=pm.join(smdata)
```

```python
df.sort_values(['similarity'],ascending=False)
```
