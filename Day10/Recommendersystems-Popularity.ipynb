```python
# movies ile ratinh kısmında movie_id aynı stün bunları birleşitreceğiz iki veriyi
```

```python
rating=pd.merge(movies,rating)
```

```python
rating.head()
```

```python
# En çok rating alan filmi bulma
```

```python
rating.title.value_counts()
```

```python
# En çok yorum alıp en çok yıldızı alanı birleştirceğzi
```

```python
import numpy as np
```

```python
mg=rating.groupby('title').agg({'rating':[np.size,np.sum,np.mean]})
```

```python
mg
```

```python
#  ençok yorum alanlardan en çok yıldızı alanlar
```

```python
pm=mg.sort_values(('rating','mean'),ascending=False)
```

```python
pm
```

```python
# 1 kişi 5 vermiş bu da tutarlı değil bu yüzden en fazla yorum yapanlar arasında 5 verenleri bulmalıyız 
```

```python
gs=pm['rating']['sum'].sum()
```

```python
gs
```

```python
pm['percentage']=pm['rating']['sum'].div(gs)*100     # Toplam ratingin yüzde kaçı hangi filme gelmiş onu hesapladık
```

```python
pm.head()
```

```python
pm['rank']=pm['percentage'].rank(ascending=False)
```

```python
pm.sort_values(by='rank',ascending=True)
```
