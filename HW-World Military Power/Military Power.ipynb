# World Military Power Projesini Clustering ile çözeceğiz.

```python
import pandas as pd
pd.set_option("display.max_columns",100)

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('World military power.xlsx')
```

```python
df.head()
```

```python

df.columns = df.iloc[0]
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df = df.drop(index=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_country = df.iloc[:, 0]        # İlk ülke adını al
df_values = df.iloc[:, 1::2]      # Tek indeksli sayısal sütunları al

df = pd.concat([df_country, df_values], axis=1)
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python

for col in df.columns[2:]:
    df[col] = df[col].astype(str).str.replace(",", "").astype(int)
```

```python
df.info()
```

```python
df.corr(numeric_only= True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Stil ayarları
sns.set_theme(style="whitegrid")
plt.rcParams["font.size"] = 12
```

```python
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Korelasyon matrisi
corr = df_numeric.corr()

plt.figure(figsize=(14, 12))
sns.heatmap(corr, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Askeri Değişkenler Arası Korelasyon Haritası", fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
km=KMeans()
```

```python
x = df.select_dtypes(include=["int64", "float64"])
```

```python
vis=KElbowVisualizer(km,k=(2,10))
vis.fit(x)
vis.show()
```

```python
model = KMeans(4)
```

```python
model = model.fit(x)
```

```python
tahmin = model.predict(x)
```

```python
tahmin
```

```python
x['cluster'] = tahmin
```

```python
silhouette_score(x , tahmin)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x["Tank Strength value"], x["defense spending budget value"], c=tahmin, cmap='viridis', s=100, alpha=0.8)
plt.colorbar(label="Küme No")
plt.xlabel("Tank Sayısı")
plt.ylabel("Savunma Bütçesi (USD)")
plt.title("K-Means Kümeleme: Tank vs Savunma Bütçesi", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data  = linkage(x , method = 'ward' , metric = 'euclidean')
```

```python
plt.figure(figsize= (25,6))
dendrogram(data)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(min_samples= 5)
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
x2=StandardScaler().fit_transform(x)
y=db.fit_predict(x2)

plt.scatter(x['Tank Strength value'],x['defense spending budget value'],c=y,cmap='autumn')
```

```python
from sklearn.cluster import AgglomerativeClustering  ##

agg_clustering = AgglomerativeClustering(n_clusters=4)
y_agg = agg_clustering.fit_predict(x)

# Plot the results
plt.scatter(x['Tank Strength value'],x['defense spending budget value'], c=y_agg, s=50, cmap='viridis')
plt.title("Agglomerative Clustering")
plt.show()
```
