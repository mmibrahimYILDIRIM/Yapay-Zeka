```python
import pandas as pd
```

```python
df=pd.read_csv('pima-indians-diabetes.csv')
```

```python
from sklearn.preprocessing import normalize,scale
```

```python
x=df.drop('Outcome',axis=1)
```

```python
x.head()    # Dataların hepsinin birimleri farklı
```

```python
# Burda makine öğrenmesi büyük rakamlara daha fazla değer verirken düşük değerlere az önem veriyor aslında burda biri daha kritik olabilir
```

```python
# unitleri aynıya getirmek için scale kullanıyoruz biri kg biri basınç hepsi farklı o yüzden bunu yapmamız gerekiyor
```

```python
yenix=scale(x)   # Bu komutla hepsi aynı unite gelmiş oldu
```

```python
yenix    # veriler dağınık o yüzden dataframe çeviriyorz
```

```python
pd.DataFrame(yenix)   #Rakamların hepsini aynı scalaya getirdi ve hepsi df datadaki değerleriyle aynı şeyi temsil ediyor mesela age 50 ile 1.42 aynı şeyi temsil ediyor
```

```python
normalized=normalize(x)
```

```python
pd.DataFrame(normalized)    # 0 la 1 arasında değerler alacak şekilde datayı düzenledik
```
