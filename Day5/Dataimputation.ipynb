```python
import pandas as pd
```

```python
df.head    # Nan yazan yerlerde data yok boş veri bunları doldurmamız lazım
```

# Doldurma Yöntemleri

## 1.Linear Regression

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
imputed_data    # Boş verileri bu şekilde doldurduk
```

```python
pd.DataFrame(imputed_data)    # Doldurulmuş verileri gördük
```

## 2.News Neighboor

```python
# Komşulara bakılıyor en yakın 5 komşunun ortalaması alınıyor
```

```python
from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer()
```

```python
pd.DataFrame(imputed_data)
```

## 3. Strateji il doldurma

```python
from sklearn.impute import SimpleImputer   # Strateji belirleyerek doldurtuyoruz ortalama orta değer vs
```

```python
imp=SimpleImputer(strategy='mean')  # median, mode gibi şeylerle doldurtabiliyoruz
```

```python
imputed_data3=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3)
```
