```python
url='https://en.wikipedia.org/wiki/Natural_language_processing'
```

```python
import requests as rq   # İnternetteki yazıları python ile okumaya yarayan paket
```

```python
yazı=rq.get(url)
```

```python
yazı.text
```

```python
from bs4 import BeautifulSoup as bs   # kötü görünümlü html kodunu okunabilir hale getiriyor
```

```python
# pypi üzerinden paketlere ulaşabilirsin
```

```python
corba=bs(yazı.text,'lxml')    # Yazıyı düzenlettik
```

```python
corba
```

```python
for p in corba.find_all('p'):
    print(p.text) # Bütün yazıları çektik NHLP için olan
```
