```python
page4=requests.get('https://www.karar.com/yazarlar/zafer-acar')
```

```python
soup=BeautifulSoup(page4.content,'html.parser')
```

```python
linkler=soup.find_all('article',{'class':'item box-shadow'})
```

```python
linkler
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    print('https://www.karar.com'+link.a.get('href'))
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    link1='https://www.karar.com'+link.a.get('href')
    baslik=link.h3.text      # Başlıkları aldık
    page4=requests.get(link1)
    soup2=BeautifulSoup(page4.content,'html.parser')
    makale=soup2.select('article.article-content')[0].text
    yazilar.append(makale)
    basliklar.append(baslik)
    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=['Title','Article'])
```

```python
df['Title']=basliklar
df['Article']=yazilar
```
